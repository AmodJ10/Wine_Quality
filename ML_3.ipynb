# Checking For Outliers

```python
pltnum = 1
sns.set(rc ={"figure.figsize":(20,15)})
for i in df:
    if pltnum<=12:
        ax = plt.subplot(3,5,pltnum)
        sns.kdeplot(df[i])
        pltnum += 1
        plt.xlabel(i,fontsize = 20)
plt.show()
```

```python
sns.boxplot(df)
```

## Removing Outliers By Trimming

```python
data_cleaned = df[df["total sulfur dioxide"] < df["total sulfur dioxide"].quantile(0.95)]
data_cleaned  = data_cleaned[data_cleaned["residual sugar"] < data_cleaned["residual sugar"].quantile(0.95)]
data_cleaned  = data_cleaned[data_cleaned["free sulfur dioxide"] < data_cleaned["free sulfur dioxide"].quantile(0.97)]
data_cleaned  = data_cleaned[data_cleaned["fixed acidity"] < data_cleaned["fixed acidity"].quantile(0.97)]
data_cleaned = data_cleaned[data_cleaned["sulphates"] < data_cleaned["sulphates"].quantile(0.97)]
data_cleaned  = data_cleaned[data_cleaned["chlorides"] < data_cleaned["chlorides"].quantile(0.97)]
```

```python
sns.boxplot(data_cleaned)
```

```python
pltnum = 1
sns.set(rc ={"figure.figsize":(20,15)})
for i in df:
    if pltnum<=12:
        ax = plt.subplot(3,5,pltnum)
        sns.kdeplot(data_cleaned[i])
        pltnum += 1
        plt.xlabel(i,fontsize = 20)
plt.show()
```

#### Now the data looks good

# Independent And Dependent Features

```python
X = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1]
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

# Model Training

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
svc_pred = svc.predict(X_test)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
score  = accuracy_score(svc_pred,y_test)
```

```python
score
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold



param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}


num_folds = 5


kf = KFold(n_splits=num_folds)


grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=kf, scoring='accuracy')


grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Best Hyperparameters:", best_params)
print("Best Score:", best_score)
```

```python
svc = SVC(C=10 , gamma="auto" , kernel="rbf")
```

```python
svc.fit(X_train,y_train)
pred = svc.predict(X_test)
score  = accuracy_score(pred,y_test)
print(score)
```
